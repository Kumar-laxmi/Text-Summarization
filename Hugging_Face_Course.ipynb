```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier([
    "I've been waiting for the Hugging Face course my whole life.",
    "I hate this so much!"
])
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about Transformers library",
    candidate_labels = ["education", "politics", "business"],
)
```

- The score generated are called as confidence levels

```python
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> model.", top_k = 2)
```

```python
ner = pipeline("ner", grouped_entities = True)
ner("My name is Kumar Laxmikant and I study at Amrita School of Engineering")
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question = "Where do I work?",
    context = "My name is Kumar Laxmikant and I work at Hugging face brooklyne"
)
```
