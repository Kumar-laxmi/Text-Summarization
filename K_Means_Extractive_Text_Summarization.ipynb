# **Read text from text file**


```python
with open('test.txt') as s:
  temp = s.readlines()
print(temp)
```

```python
text = ""
for line in temp:
  text = text + line 
```

```python
text
```

# **Importing Libraries**

```python
#importing the libraries

import nltk
import numpy as np
import re
from gensim.models import Word2Vec
from sklearn.cluster import KMeans
from scipy.spatial import distance
nltk.download('punkt')   # one time execution
nltk.download('stopwords')  # one time execution
from nltk.corpus import stopwords
```

# **Split text into sentences**
<p>
nltk.tokenize.punkt module which is pre-trained to model abbreviations, characters and punctuations to mark the beginning and end of sentences.
</p>

```python
#sentence tokenization

from nltk.tokenize import sent_tokenize
sentence = sent_tokenize(text)

for sen in sentence:
  print("- {}".format(sen),end="\n")
```

# **Generate clean text**
- Replacing every element that is not a letter by space
- Transforming capital letters into lowercase
- Removing stopwords which are commonly used words like articles, prepositions which do not carry any information and donâ€™t help us towards our goal of summarization.

```python
# cleaning the sentences

corpus = []
for i in range(len(sentence)):
    sen = re.sub('[^a-zA-Z]', " ", sentence[i])  
    sen = sen.lower()                            
    sen=sen.split()                         
    sen = ' '.join([i for i in sen if i not in stopwords.words('english')])   
    corpus.append(sen)
    print("\n")
    print("Pass = {}".format(i+1))
    print("- {}".format(sen))
```

# **Vector representation of sentences**
<p>
Word2Vec are basically a form of vector representation of words to bridge the gap between human understanding of language to that of a machine.
</p>

```python
#creating word vectors

all_words = [i.split() for i in corpus]

for words in all_words:
  print("- {}".format(words))
```

```python
n=300

model = Word2Vec(all_words, min_count=1,size= n)
model
```

<p>
For creating sentence vectors, we will grab vector representation for all the constituent words in a sentence and then take average to arrive at an amalgamated vector.
</p>

```python
# creating sentence vectors

sen_vector=[]
for i in corpus:
    
    plus=0
    for j in i.split():
        plus+=model.wv[j]
    plus = plus/len(plus)
    
    sen_vector.append(plus)
    print("- {}".format(plus))
```

# **Clustering**
<p>
After forming sentence vectors we will model and perform clustering to group sentence embeddings into a pre-defined number of clusters which is equal to the desired number of sentences we want in our summary.
</p>

```python
#performing k-means  
    
n_clusters = int(input("Number of sentences: "))
kmeans = KMeans(n_clusters, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(sen_vector)
y_kmeans
```

```python
#finding and printing the nearest sentence vector from cluster centroid


my_list=[]
for i in range(n_clusters):
    my_dict={}
    
    for j in range(len(y_kmeans)):
        
        if y_kmeans[j]==i:
            my_dict[j] =  distance.euclidean(kmeans.cluster_centers_[i],sen_vector[j])
    min_distance = min(my_dict.values())
    my_list.append(min(my_dict, key=my_dict.get))
    print("- {}".format(min(my_dict, key=my_dict.get)))
```

```python
for key, value in my_dict.items():
  print("{} : {}".format(key,value))
```

```python
print(my_list)
```

```python
print(y_kmeans)  
```

```python
for i in sorted(my_list):
    print("- {}".format(sentence[i])) 
```
